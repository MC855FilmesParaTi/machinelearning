```python
import pandas as pd
import numpy as np

from scipy.sparse import csc_matrix
```

```python
data_dir = '../data/processed/'
```

```python
rating_df = pd.read_csv(data_dir + 'train_rating.csv')
rating_df.shape
```

```python
reduced = rating_df.sample(5000, random_state=42).sort_values(by='movieId')
reduced.head()
```

```python
reduced = reduced[['userId','movieId','rating']]
reduced.head()
```

```python
reduced.userId.unique().shape
```

```python
reduced.movieId.unique().shape
```

```python
M = reduced.pivot_table( values='rating', index='movieId', columns='userId' )
M.fillna(0, inplace=True)
M
```

```python
Sparse = csc_matrix( M )
Sparse
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(
    # Quantidade de recomendações
    n_neighbors=6,
    # Algoritmo
    algorithm='brute',
    # metrica de distancia
    metric='minkowski'
)

model.fit(Sparse)
```

```python
reduced.movieId.sort_values()
```

```python
movies_df = pd.read_csv(data_dir + 'movie.csv')
movies_df = movies_df[movies_df.movieId.isin(reduced.movieId)]
movies_df[['movieId','title']].sample(5)
```

```python
M[M.index == 51925]
```

```python
sample = M[M.index == 51925].values
sample.shape
```

```python
distances, indexes = model.kneighbors(sample)
indexes = indexes[0]
indexes
```

```python
movies_Id_rec = np.asarray(M.iloc[indexes].index)
movies_Id_rec
```

```python
movies_df[movies_df.movieId.isin(movies_Id_rec)]
```
